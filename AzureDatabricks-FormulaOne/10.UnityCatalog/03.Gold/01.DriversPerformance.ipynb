```sql
DROP TABLE IF EXISTS formulaone_dev.gold.drivers_win;
```

```sql
CREATE TABLE IF NOT EXISTS formulaone_dev.gold.drivers_win
AS
SELECT
drivers.forename
,drivers.surname
,COUNT(1) AS overall_win
FROM
formulaone_dev.silver.drivers drivers JOIN formulaone_dev.silver.results results 
ON drivers.driverId = results.driverId
WHERE results.position = 1
GROUP BY ALL;
```

```sql
SELECT * FROM formulaone_dev.gold.drivers_win ORDER BY overall_win DESC;
```
