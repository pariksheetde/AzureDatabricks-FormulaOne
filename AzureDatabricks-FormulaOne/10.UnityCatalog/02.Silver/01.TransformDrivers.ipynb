#### CREATE `DRIVERS` TABLE AT `SILVER` SCHEMA
- ADD `_metadata.file_path` TO IDENTIFY THE SOURCE

```sql
DROP TABLE IF EXISTS formulaone_dev.silver.drivers;
```

```sql
CREATE TABLE IF NOT EXISTS formulaone_dev.silver.drivers
AS
  SELECT 
    driverId
    ,driverRef
    ,number
    ,code
    ,name.forename AS forename
    ,name.surname AS surname
    ,dob
    ,nationality
    ,CURRENT_TIMESTAMP() AS ingestion_timestamp
    ,_metadata.file_path AS file_path
   FROM formulaone_dev.bronze.drivers;
```

```sql
SELECT * FROM formulaone_dev.silver.drivers;
```
